# Python Functions

In Python, a function is a group of related statements that performs a specific task.

Functions help break our program into smaller and modular chunks. As our program grows larger and larger, functions make it more organized and manageable.

Furthermore, it avoids repetition and makes the code reusable.

## Syntax of Function


```python
def function_name(parameters):
    """docstring"""
    statements(s)
```

1. Keyword def that marks the start of the function header.

2. A function name to uniquely identify the function.

3. Parameters (arguments) through which we pass values to a function. They are optional.

4. A colon (:) to mark the end of the function header.

5. Optional documentation string (docstring) to describe what the function does.

6. One or more valid python statements that make up the function body. 

7. Statements must have the same indentation level (usually 4 spaces).

8. An optional return statement to return a value from the function.

## Parameters or Arguments

The terms parameter and argument can be used for the same thing: information that are passed into a function.

From a function's perspective:

A parameter is the variable listed inside the parentheses in the function definition.

An argument is the value that is sent to the function when it is called.

```python
def minuss(height, width):
    return height-width

print('positional arguments (10, 100) : ', minuss(10,100))
print('positional arguments (100, 10) : ', minuss(100, 10))
print('keyword arguments (width = 100, height = 10) : ', minuss(width = 100, height = 10))
print('positional & keyword arguments (100, width = 10) : ', minuss(100, width = 10))
```

```python
try : 
    minuss(10, height=100)
except Exception as e:
    #pass
    print(e)
```

## Unpacking With the Asterisk Operators: * & **


\*args and **kwargs allow you to pass multiple arguments or keyword arguments to a function.

```python
def my_sum(*integers):
    result = 0
    print(integers)
    for x in integers:
        result += x
    return result

print(my_sum(1, 2, 3))
```

```python
def string_sum(**words):
    result = ""
    print(words)
    for arg in words.values():
        result += arg
    return result

print(string_sum(a="Real", b="Python", c="Is", d="Great", e="!"))
```

```python
def g(a, b, *args, **kwargs):
    print(a,b)
    print(args)
    print(kwargs)

g(1,2,3,4,c=5,d=6)
g(10,3,4,c=5,d=6)
g(100, b=20, c=7)
```

```python
def h(a, b, c):
    print(a, b, c)

a = (1, 2, 3)
h(*a)
```

```python
dargs={'a':1, 'b':2, 'c':3}
h(**dargs)
```

```python
args = (1,2)
dargs = {'c':3}
h(*args, **dargs)
```

```python
def my_sum(*args):
    result = 0
    for x in args:
        result += x
    return result

list1 = [1, 2, 3]
list2 = [4, 5]
list3 = [6, 7, 8, 9]

print(my_sum(*list1, *list2, *list3, 10))
```
