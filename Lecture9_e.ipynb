# Dictionary

Dictionaries are written with curly brackets, and have keys and values.

A dictionary is a collection which is ordered*, changeable and does not allow duplicates.

Dictionary items are presented in key:value pairs, and can be referred to by using the key name.

Dictionary keys are unchangeable and hashable. 

Constructor : dict( )

```python
sample_dict = {}
sample_dict['one']=1
sample_dict['two']=2
print(sample_dict)
```

```python
sample_dict_2 = dict(three=3, four=4)
sample_dict_3 = {'five':5, 'six':6}
print(sample_dict_2)
print(sample_dict_3)
```

```python
def adds(a,b):
    return (a+b)
def subs(a,b):
    return (a-b)

c = adds(4,5)
d = subs(4,5)
print(c,d)

```

```python
actions = {0:adds, 1:subs}
print(actions[0](4,5))
print(actions[1](4,5))
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
```

You can access the items of a dictionary by referring to its key name, inside square brackets:



```python
sample_dict_2['three']
```

To determine if a specified key is present in a dictionary use the in keyword:

```python
if "four" in sample_dict_2:
  print("Yes, 'four' is one of the keys in the sample_dict_2 dictionary")
```

### keys()
Return a new view of the dictionary’s keys. 

### values()
Return a new view of the dictionary’s values.

### items()
Return a new view of the dictionary’s items ((key, value) pairs).

```python
sample_dict.keys()
```

```python
type(sample_dict.keys())
```

```python
list(sample_dict.keys())
```

```python
print(sample_dict.values())
print(list(sample_dict.values()))
```

```python
print(sample_dict.items())
print(list(sample_dict.items()))
```

You can change the value of a specific item by referring to its key name:



```python
sample_dict['one']=1000
```

```python
sample_dict
```

```python
sample_dict.values()
```

Adding an item to the dictionary is done by using a new index key and assigning a value to it:



```python
sample_dict['color'] = {'red':(255,0,0), 'green':(0,255,0), 'blue':(0,0,255)}
```

```python
sample_dict
```

The pop() method removes the item with the specified key name:



```python
sample_dict.pop('two')
```

```python
sample_dict
```

The del keyword removes the item with the specified key name:



```python
del sample_dict['one']
sample_dict
```

```python
sample_dict['color']['blue'][2]
```

### Loop Through a Dictionary

```python
for i in sample_dict_3:
    print(i)

print('================')
for i in sample_dict_3.values():
    print(i)

print('================')
for i, j in sample_dict_3.items():
    print(i,j)
```

You cannot copy a dictionary simply by typing dict2 = dict1, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

```python
dict1 = dict(sample_dict)
dict2 = dict1
print(dict1, dict2)
```

```python
dict1[(12,1)] = 1729
```

```python
print(dict1)
print(dict2)
print(sample_dict)
```

```python
dict3 = dict1.copy()
print(dict3)
print(dict1)
```

```python
dict3[(10,9)]=1729
```

```python
dict3
```

```python
print(dict1)
print(dict2)
print(dict3)
print(sample_dict)
```
