```
import numpy as np
import math
def y_value(x):
    y=[]
    for i in x:
        y.append(math.sin(i))
    return y

x_value = np.arange(1,10,math.pi/100)
y = y_value(x_value)
```

```
x_value, y
```

```
plt.plot(x_value,y, 'g', label='Sine Plot')
plt.legend(loc = 'upper right',fontsize=10)
plt.title('y=sin(x)',fontsize=20)
plt.xlabel('Radian',fontsize=20)
plt.ylabel('sin(x)',fontsize=20)
plt.grid()
plt.text(3,0.5, "This Point")
plt.tight_layout()
```

```
recipes = pd.read_csv('recipes_muffins_cupcakes.csv')
recipes
```

```
figure = plt.figure()
axes1 = figure.add_subplot(1,2,1)
axes2 = figure.add_subplot(1,2,2)
axes1.plot(recipes.Flour)
axes2.plot(recipes.Milk)
```

```
axes1.set_title('Flour')
axes2.set_title('Milk')
figure.suptitle('recipes')
figure
```

```
recipes.Sugar.plot(kind='hist', bins=10, color='blue')
```

```
recipes.plot(x='Milk', y='Sugar', kind='scatter', c='blue')
```

```
import seaborn as sns
sns.pairplot(recipes, hue='Type') # hue 에 'Type'을 지정하면, Type에는 두 종류 데이터가 있어서 두 개의 색으로 표출
```
